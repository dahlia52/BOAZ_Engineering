### TCP와 UDP

- **OSI 7계층 모델** : 국제 표준화 기구에서 네트워크간의 연결에 있어 호환성 결여를 막기 위해 규정한 모델 

- **TCP/IP 모형** : OSI 모델을 기반으로 실제 인터넷에서 컴퓨터간의 정보 통신 시 쓰이는 통신규약(프로토콜)의 모음으로 각 계층은 담당하는 위치마다 처리하는 역할을 구분함. (OSI보다 더 실용적)

### 소켓 (socket)

- **소켓** : 네트워크 통신에서 데이터 송수신의 접점 

  - TCP/UDP를 이용하여 데이터를 주고 받는 창구 역할을 함. 
  
  - 서버와 클라이언트가 서로 통신할 수 있도록 양쪽에 생성되는 링크 단자

- **HTTP 통신** : 클라이언트가 요청을 보내고 요청에 따라 서버가 응답하는 **단방향 통신**

- **소켓 통신** : 서버-클라이언트간 데이터를 주고 받는 **양방향 통신**

  - 소켓은 보통 지속적으로 연결을 유지하며 실시간으로 데이터를 주고 받음 (Real-time Networking) 

  - 클라이언트가 서버에게 직접 요청하지 않아도 서버에서 실시간 정보를 클라이언트에게 보여줌.
    
    ex. 주식, 채팅 


### 소켓 실행 흐름

- 소켓통신을 위해서는 **클라이언트(client)와 서버(Server) 모두 소켓을 생성**해야 함.

Client

1. 소켓 생성 `socket()`

2. 서버 소켓과 연결 `connect()`

3. 서버에 메세지 송신, 수신

리눅스

- 리눅스 탄생 전에는 유닉스(Unix)라는 운영체제가 널리 사용됨. 유닉스는 현재까지도 비싼 비용을 지불해야 사용할 수 있음. 리눅스가 **유닉스를 대체**하게 됨.

#### 디렉터리 명령어

- **pwd** (Print Working Directory) : 현재 어느 디렉토리에 있는지 알려줌 

- **cd [디렉토리명]** (Change Directory) : 다른 디렉토리로 이동 (..은 상위 디렉토리로의 이동을 의미함)

- **ls** (list) : 현재 디렉토리에 있는 파일과 하위 디렉토리 목록을 보여줌

- **mv [원본이름] [대상이름]** (move) : 원본 파일을 대상 파일로 이동시킴

- **cp [원본이름] [대상이름]** (copy) : 원본 파일을 대상 파일로 복사함. 옵션 -r을 붙이면 디렉토리를 복사할 수 있음.

- **rm [파일명]** (remove) : 파일을 삭제함. 옵션 -r을 붙이면 디렉토리를 삭제할 수 있음.

- **mkdir [디렉토리명]** (make directory) : 새로운 디렉토리를 만듦

- **cat [파일명]** (catch) : 파일 내용 출력

$ ls 명령어

```python
# 루트 디렉터리의 내용 리스트
ls /
# 홈 디렉터리의 내용 리스트
ls ~
# 숨겨진 파일을 포함한 모든 파일을 리스트로 표시(all)
ls -a
# 파일의 크기를 K 바이트 단위로 출력(size)
ls -s
# 파일의 상세 정보(이름, 종류, 접근 권한, 소유자, 크기, 수정시간 등)을 출력(long)
ls -l
```

### 파일의 권한 변경 (pem 파일의 권한 변경)

- **chmod** (Change Mode) : 파일의 권한을 바꿈. 이 명령어를 사용하기 위해서는 파일에 접근할 수 있는 권한이 필요하기 때문에 파일 소유자이거나 루트 권한이 필요함.

- chmod 명령어 옵션

  영문자 옵션

  - u : 소유자
  - g : 소유 그룹
  - o : 제 3자

  - +r / -r : 읽기 권한 추가, 삭제
  - +w / -w : 쓰기 권한 추가 또는 삭제
  - +x / -x : 실행 권한 추가 또는 삭제

  ex. $ chmod go+rwx Image : Image 디렉토리의 소유그룹과 제 3자에게 읽기, 쓰기, 실행 권한을 추가

  ex. $ chmod o-xw Image : Image 디렉토리의 제 3자에게  쓰기, 실행 권한을 삭제

  숫자 옵션 : 권한마다 숫자를 지정해서 더해주는 방법으로 권한 설정
  - R : 4
  - W : 2
  - X : 1

  ex. $ chmod 755 Image : 소유자는 7(4+2+1)으로 읽기, 쓰기, 실행 권한 부여, 소유 그룹과 제 3자는 5(4+1)로 읽기, 실행 권한만 부여
