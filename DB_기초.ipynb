## DBMS (DataBase Management System)

- **데이터** : 관찰의 결과로 나타난 정량적 혹은 정성적인 실제 값

- **정보** : 데이터에 의미를 부여한 것

- **데이터베이스**: 조직에 필요한 정보를 얻기 위해 논리적으로 연관된 데이터를 모아 구조적으로 통합해 놓은 것

#### DBMS 발전 과정



- **파일시스템**(0세대) : 컴퓨터에서 파일이나 자료를 쉽게 발견 및 접근할 수 있도록 보관 또는 조직하는 체제

- **계층형 HDBMS**(1세대) : 계층적인 형태의 DBMS, 초기 세팅이 변화하면 이에 대처하기 힘들어 잘 사용X

- **네트워크형 NDBMS**(2세대) : 구성과 설계가 복잡하고 데이터 종속성을 해결하지 못함

- **관계형 RDBMS** (3세대) : 테이블과 테이블 간의 관계를 기반으로 하는 가장 범용적인 데이터베이스 관리 시스템

- **객체지향 ODBMS** (4세대) : 정보를 객체의 형태로 표현하는 DBMS

- **NoSQL** (5세대) : 데이터 간의 관계를 설정하지 않고 유연한 테이블 스키마를 가짐.

#### DBMS의 특징



- **실시간 접근성**

  - DB는 실시간으로 서비스됨.
  - 사용자가 데이터를 요청하면 수 초 내에 결과를 전송함.

- **계속적인 변화**

  - DB에 저장된 내용은 어느 한 순간의 상태를 나타내지만, 데이터 값은 시간에 따라 항상 바뀜. DB는 삽입, 삭제, 수정 등의 작업을 통하여 바뀐 데이터 값을 저장함.

- **동시 공유**

  - DB는 서로 다른 업무 또는 여러 사용자에게 동시에 공유됨.
  - DB에 접근하는 프로그램이 여러 개 있음.

- **내용에 따른 참조**
  - DB에 저장된 데이터는 데이터의 물리적인 위치가 아니라 데이터 값에 따라 참조됨.

## RDBMS (Relational Database Management System)

**관계형 데이터베이스 (테이블)**: 모든 데이터를 2차원 형태의 테이블로 표현

- 행(row) / 열(column)

- 값 (value)
  - 각각의 행과 열에 대응하는 값을 가지고 있음.
  - 값은 열의 타입에 맞는 값이어야 함.
- 키 (key)
  - 테이블에서 행의 식별자로 이용되는 열
  - 기본키, 슈퍼키, 외래키

- 관계 (relationship): 일대일(one-to-one), 일대다(one-to-many), 다대다(many-to-many)

- 스키마 (schema)
  - 테이블 디자인을 위한 청사진
  - 테이블의 각 열에 대한 항목과 타입, 기본키와 외래키 기재


image.png

#### RDBMS의 특징

- 데이터의 분류, 정렬, 탐색 속도가 빠름
- 신뢰성이 높고, 데이터의 무결성 보장
- 기존에 작성된 스키마를 수정하기 어려움

장점

- 정해진 스키마에 따라 데이터를 저장해서 명확한 데이터 구조 보장
- 각 데이터를 중복 없이 한 번씩 저장 가능


단점
- 테이블 간 관계성 : 시스템이 커지면 복잡한 쿼리 생성
- 스키마로 인해 데이터가 유연하지 않음 → 스키마 변경 시 복잡함.

#### 무결성 제약 조건

**데이터의 무결성**: 데이터베이스에 저장된 데이터의 일관성과 정확성을 지키는 것

- **도메인 무결성 제약 조건**
  - relation 내의 튜플들이 각 속성의 도메인에 지정된 값만을 가져야 한다는 조건
  - SQL문에서 데이터 형식(type), null, 기본 값(default), check 등을 사용하여 지정

- **개체 무결성 제약 조건** (기본키 제약 조건)
  - 기본키는 Null 값을 가져서는 안 되며 relation 내에 오직 하나의 값만 존재해야 함.

- **참조 무결성 조건**
  - relation의 외래키는 부모 relation의 기본키와 도메인이 동일해야 하며, 자식 relation의 값이 변경될 때 부모 relation의 제약을 받음.

### 트랜잭션

**Transaction (트랜잭션)**

- DBMS가 데이터를 사용하는 작업 (프로그램) 단위

- 단일 / 여러 개의 SQL문을 순차적으로 수행하여 다룸.


#### 트랜잭션의 특징 (ACID)

- **Atomicity** (원자성)
  - 시스템에서 한 트랜잭션의 연산들이 모두 성공하거나 실패
  - 트랜잭션 중간에 작업이 잘못되면 회복 알고리즘(recovery)을 이용하여 변경 내용을 취소, 자의적으로 트랜잭션을 취소하는 명령어(rollback)

- **Consistency** (일관성)
  - 테이블이 생성될 때, 명령문의 무결성 제약조건을 통해 명시

- **Isolation** (고립성)
  - 모든 트랜잭션은 다른 트랜잭션으로부터 독립되어야 함.
  - 변경 중인 임시 데이터에 대해 다른 트랜잭션이 읽거나 쓰려고 할 때 제어하는 작업이 필요

- **Duration** (지속성)
  - 하나의 트랜잭션이 정상적/부분적으로 완료되었을 때, 해당 데이터는 데이터베이스에 기록되어야 함.
  - 로그 데이터베이스를 통해 문제 발생 시, 복구 작업 수행. 시스템이 멈춰도 트랜잭션 수행으로 변경된 내용은 디스크에 기록

## SQL 

#### DDL (Data Definition Language) : 데이터 정의어

데이터를 생성, 수정, 삭제하는 등 데이터의 전체의 골격을 결정하는 역할을 하는 언어

- **CREATE** : 테이블 생성

- **ALTER** : 데이터 변경

- **DROP** : 데이터 삭제 (데이터 구조 전체 삭제)



#### DML (Data Manipulation Language) : 데이터 조작어

정의된 데이터베이스에 입력된 레코드를 조회하거나 수정, 삭제하는 등의 역할을 하는 언어

- **SELECT** : 데이터 조회 / 검색

- **INSERT** : 데이터 삽입

- **UPDATE**: 데이터 업데이트

- **DELETE**: 데이터 삭제 (특정 데이터값 삭제)

#### DCL (Data Control Language) : 데이터 제어어

데이터베이스에 접근하거나 객체에 권한을 주는등의 역할을 하는 언어

- **GRANT** : 권한 허가
- **REVOKE** : 권한 취소

#### TCL (Ttansaction Control Language) : 트랜잭션 제어어



- **COMMIT** : 트랜잭션의 작업을 저장
- **ROLLBACK** : 트랜잭션의 작업을 취소, 원래대로 복구
- **LOCK** : 트랜잭션의 동시성 제어를 위한 명령어
- **SAVEPOINT** : 현재의 트랜잭션을 분할하여 작업 수행할 수 있도록 돕는 명령어

## NoSQL

**NoSQL** (Not Only SQL)

- 인터넷의 활성화, 소셜네트워크 서비스의 등장 → 비정형 데이터 활용도 증가

- ACID 특성을 제공하지 않지만 확장성이나 성능이 뛰어나  비관계형, 분산형 데이터 베이스 시스템으로 적합

- 관계형 모델 사용 x, 테이블 간의 조인 기능이 없음.

- 대부분 여러 대의 데이터베이스 서버를 묶어 하나의 데이터베이스를 구성

- 관계형 데이터베이스에서 지원하는 ACID 특성 미보장

- 스키마와 속성들을 다양한 형태로 정의 가능 -> 초고용량 데이터 처리 성능에 특화, 비관계형 데이터 저장소에 저장을 위한 분산 저장 시스템


#### NoSQL 종류

- **Document Store**

  - 데이터 및 메타 데이터를 데이터 베이스 내부의 JSON 기반 문서에 계층적으로 저장
  - 콘텐츠를 기반으로 문서를 검색하는 API 또는 쿼리 언어

- **KEY-Value Store**

  - 가장 단순한 데이터베이스인 키-값 모음으로 표현
  - 키를 사전식 순서로 유지하는 불연속적 정렬 → 선택적 키 범위를 효율적으로 검색 가능

- **Wide-Column Store**
  - RDBMS의 속성을 계층적으로 표현
  - 관련 데이터를 단일 열 내에 중첩된 키/값 세트로 저장

- ** Graph Store**
  - 데이터를 Node, Edge 형태로 그래프 구조에 저장
  - ex) SNS, 대중교통 링크, 도로 지도
